```python
My_Data = Temiz_Veri
My_Data = My_Data[My_Data.year != 1970]
```

```python
My_Data.groupby("year").mean()["price"]
```

```python
My_Data = My_Data.drop("transmission",axis=1)
My_Data
```

```python
y = My_Data["price"].values
x = My_Data.drop("price",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300)
```

```python
kayipVeri = pd.DataFrame(model.history.history)
```

```python
kayipVeri.plot()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
tahminDizisi = model.predict(x_test)
```

```python
mean_absolute_error(y_test,tahminDizisi)
```
