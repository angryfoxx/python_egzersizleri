```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yas_listesi = np.array([10,20,25,30,40,50,55,60,70,80])
```

```python
kilo_listesi = np.array([30,70,90,55,70,80,120,65,130,70])
```

```python
plt.plot(yas_listesi,kilo_listesi,"g")
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Kilo'nun yaşa göre grafiği")
#plt.show()
```

```python
dizi_1 = np.linspace(1,10,20)
dizi_2 = dizi_1 ** 3
```

```python
plt.subplot(1,2,1)
plt.plot(dizi_1,dizi_2,"g--")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.title("1. Satır 1. Sütun")
plt.subplot(1,2,2)
plt.plot(dizi_2,dizi_1,"r--")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.title("1. Satır 2. Sütun")
```

```python
my_figure = plt.figure()
my_axes = my_figure.add_axes([0.1,0.1,0.8,0.8])
my_axes.plot(dizi_1,dizi_2,"g--")
my_axes.set_xlabel("X Ekseni")
my_axes.set_ylabel("Y Ekseni")
my_axes.set_title("Grafik Başlığı")
```

```python
my_figure = plt.figure()
my_axes_1 = my_figure.add_axes([0.1,0.1,0.9,0.9])
my_axes_2 = my_figure.add_axes([0.2,0.52,0.35,0.35])

my_axes_1.plot(dizi_1,dizi_2,"g--")
my_axes_1.set_xlabel("X Ekseni")
my_axes_1.set_ylabel("Y Ekseni")
my_axes_1.set_title("Büyük Grafik")

my_axes_2.plot(dizi_2,dizi_1,"b--")
my_axes_2.set_xlabel("x Ekseni")
my_axes_2.set_ylabel("y Ekseni")
my_axes_2.set_title("Küçük Grafik")
```

```python
(benimFigure, benimEksen) = plt.subplots(nrows=1, ncols=2)
for i in benimEksen:
    i.plot(dizi_1,dizi_2,"g--")
    i.set_xlabel("X Ekseni")
    i.set_ylabel("X Ekseni")
    i.set_title("Burası Bir Başlık")
plt.tight_layout()
```

```python
figur = plt.figure() #dpi = 200
eksen = figur.add_axes([0.1,0.1,0.9,0.9])
eksen.plot(dizi_1, dizi_2, "g", label = "dizi_1 ^ 3")
eksen.plot(dizi_1, dizi_1 ** 2 , "r", label = "dizi_1 ^ 2")
eksen.set_xlabel("X Ekseni")
eksen.set_ylabel("Y Ekseni")
eksen.set_title("Burası Başlık")
eksen.legend(loc=2)
```

```python
figur.savefig("benimfigur.png",dpi=200)
```
